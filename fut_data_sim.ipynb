```
import random as rd
import pandas as pd
```

```
# dictionary containing teams and there stats, offence multiplier and defence multiplier
#{id:["nome",mult_off,multi_def]}
teams_dict ={
    "c1":["FC Croquete",1.1,1.2],
    "c2":["AD Passarinho",1.0,1.1],
    "c3":["UD Febra",1.3,1.0],
    "c4":["AR Pauliteiros",1.0,1.0],
    "c5":["AS Bife",1.1,1.0],
    "c6":["Javali",1.0,1.3],
    "c7":["Escuteiros",1.0,1.1],
    "c8":["AS Zebreus",1.2,1.2],
    "c9":["AS Tonys",1.4,0.9],
    "c10":["Batatas",1.1,0.8],
    "c11":["Pão de Forma Club",0.9,0.8],
    "c12":["Kangurus Amestrados",0.7,1.2],
    "c13":["Pardelho ADR",1.5,0.7],
    "c14":["Camionista FC",0.8,1.5],
    "c15":["AD Bombeiros",1.6,0.5],
    "c16":["Club 404",1.1,1.1],
    "c17":["AS Falidos",0.9,0.5],
    "c18":["Espatula",1.6,1.5],
    "c19":["Quimboio",1.5,1.7],
    "c20":["Sete&Meio",1.7,1.4]
}
```

```
#Simulate league

#dictionary, rank_dict structure:
#"id":[club_name, n_wins, n_draw, n_losses, goals_scored, goals_suffered, goals_net, points, games]

def sim_league():

  rd.seed(None,2)

  # dictionary to introduce league result values
  rank_dict ={
      "c1":["FC Croquete",0,0,0,0,0,0,0,0],
      "c2":["AD Passarinho",0,0,0,0,0,0,0,0],
      "c3":["UD Febra",0,0,0,0,0,0,0,0],
      "c4":["AR Pauliteiros",0,0,0,0,0,0,0,0],
      "c5":["AS Bife",0,0,0,0,0,0,0,0],
      "c6":["Javali",0,0,0,0,0,0,0,0],
      "c7":["Escuteiros",0,0,0,0,0,0,0,0],
      "c8":["AS Zebreus",0,0,0,0,0,0,0,0],
      "c9":["AS Tonys",0,0,0,0,0,0,0,0],
      "c10":["Batatas",0,0,0,0,0,0,0,0],
      "c11":["Pão de Forma Club",0,0,0,0,0,0,0,0],
      "c12":["Kangurus Amestrados",0,0,0,0,0,0,0,0],
      "c13":["Pardelho ADR",0,0,0,0,0,0,0,0],
      "c14":["Camionista FC",0,0,0,0,0,0,0,0],
      "c15":["AD Bombeiros",0,0,0,0,0,0,0,0],
      "c16":["Club 404",0,0,0,0,0,0,0,0],
      "c17":["AS Falidos",0,0,0,0,0,0,0,0],
      "c18":["Espatula",0,0,0,0,0,0,0,0],
      "c19":["Quimboio",0,0,0,0,0,0,0,0],
      "c20":["Sete&Meio",0,0,0,0,0,0,0,0]
  }

  weekly_schedule = []

  # loops to simulate games between al teams, home and away games
  for home in range(1,1+len(teams_dict)):
    
    for away in range(1,1+len(teams_dict)):
      goals_team_home = 0
      goals_team_away = 0
      
      # number os rounds per games
      play_rounds = 6
      
      # loop for simulatinn games
      for i in range(play_rounds):

        # test if the team are the sames, if true skips game
        if home == away:
          break

        # chooses who attack each round
        attacker_choice = rd.randint(0,1)
        
        # games simulation
        if attacker_choice == 1: 
          teamHome_s = rd.randint(0,100) * teams_dict.get(f"c{home}")[1]
          teamAway_s = rd.randint(0,100) * teams_dict.get(f"c{away}")[2]
        else:
          teamHome_s = rd.randint(0,100) * teams_dict.get(f"c{home}")[2]
          teamAway_s = rd.randint(0,100) * teams_dict.get(f"c{away}")[1]

        #decides if goal is scored and by who or draw  
        if teamHome_s > teamAway_s + 40:
          goals_team_home += 1
        elif teamHome_s < teamAway_s - 40:
          goals_team_away += 1
      
      # creates weekly schedule
        # test if the team are the sames, if true skips game
      if home == away:
        continue
      else:
        # weekly schedule format: [home team name, goals scored, away team name, goals scored]
        result = [teams_dict.get(f"c{home}")[0],goals_team_home, teams_dict.get(f"c{away}")[0],goals_team_away]
        weekly_schedule.append(result)

      # intriduces values into de league result info
      # n_wins[1], n_draw[2], n_losses[3],  points[7], Games[8]

        # test if the teams are the same, if true skips game
      if home == away:
        continue

        # legend: n_wins[1], n_draw[2], n_losses[3],  points[7], Games[8]
        # number in [ ] represents de label of information inserted (see legend above)

        #home win
      elif goals_team_home > goals_team_away:
        rank_dict[f"c{home}"][1] += 1
        rank_dict[f"c{away}"][3] += 1
        rank_dict[f"c{home}"][7] += 3
        rank_dict[f"c{home}"][8] += 1
        rank_dict[f"c{away}"][8] += 1

        #away win
      elif goals_team_home < goals_team_away:
        rank_dict[f"c{home}"][3] += 1
        rank_dict[f"c{away}"][1] += 1
        rank_dict[f"c{away}"][7] += 3
        rank_dict[f"c{home}"][8] += 1
        rank_dict[f"c{away}"][8] += 1

        #draw
      elif goals_team_home == goals_team_away:
        rank_dict[f"c{home}"][2] += 1
        rank_dict[f"c{away}"][2] += 1
        rank_dict[f"c{home}"][7] += 1
        rank_dict[f"c{away}"][7] += 1
        rank_dict[f"c{home}"][8] += 1
        rank_dict[f"c{away}"][8] += 1

      #goals_scored[4], goals_suffered[5], goals_net[6],
        #scored
      rank_dict[f"c{home}"][4] += goals_team_home
      rank_dict[f"c{away}"][4] += goals_team_away

        #sufered
      rank_dict[f"c{home}"][5] += goals_team_away
      rank_dict[f"c{away}"][5] += goals_team_home

        #goal net
      rank_dict[f"c{home}"][6] += (goals_team_home - goals_team_away)
      rank_dict[f"c{away}"][6] += (goals_team_away - goals_team_home)

  
  print(weekly_schedule) #fazer-------
  
  # creates a pandas dataframe based on the league information
  data_frame_league = pd.DataFrame.from_dict(rank_dict, orient='index',
                                             columns=['Club','V','D','L','Goals+','Goals-','Goals net','Points','Games'])
  
  # sorts data_frame_league based primary on Points and secondary on Goals net
  sorted_data_frame_league = data_frame_league.sort_values(by = ['Points','Goals net'], ascending = False)
  

sim_league()
  
```

```
#Simulates 1 match with 2 random teams

def sim_1_match():
  rd.seed(None,2)

  play_rounds = 6

  goals_team_home = 0
  goals_team_away = 0
  teamHome_s = 0
  teamAway_s = 0

  t1 = 0
  t2 = 0
  while t1 == t2:
    t1 = rd.randint(1,20)
    t2 = rd.randint(1,20)

  for i in range(play_rounds):

    attacker_choice = rd.randint(0,1)
    if attacker_choice == 1: 
      teamHome_s = rd.randint(0,100) * teams_dict.get(f"c{t1}")[1]
      teamAway_s = rd.randint(0,100) * teams_dict.get(f"c{t2}")[2]
    else:
      teamHome_s = rd.randint(0,100) * teams_dict.get(f"c{t1}")[2]
      teamAway_s = rd.randint(0,100) * teams_dict.get(f"c{t2}")[1]
    
    if teamHome_s > teamAway_s + 40:
      goals_team_home += 1
    elif teamHome_s < teamAway_s - 40:
      goals_team_away += 1
    else:
      continue

  print(teams_dict[f"c{t1}"][0], "Home", goals_team_home)
  print(teams_dict[f"c{t2}"][0], "Away", goals_team_away)

sim_1_match()
```
